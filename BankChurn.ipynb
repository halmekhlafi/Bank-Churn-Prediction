```python
min_age = df['Age'].min()
max_age = df['Age'].max()

bins = np.linspace(18, max_age, 6)  # Starting at 18, so we have 5 bins

labels = [f'{int(bins[i])}-{int(bins[i+1]-1)}' for i in range(len(bins)-1)]

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

df['Age'] = df['Age'].astype(int)

print("Data with Age Groups:")
print(df)

age_group_counts = df['Age_Group'].value_counts()

print("\nCount of values in each Age Group:")
print(age_group_counts)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Defining features and target
X = df.drop('Exited', axis=1)
y = df['Exited']

# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)  
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
```

```python
svm = SVC(kernel='rbf', probability=True)  
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
```

```python
models = {
    "Random Forest": rf_pred,
    "K-Nearest Neighbors": knn_pred,
    "Support Vector Machine": svm_pred
}

for name, pred in models.items():
    print(f"----- {name} -----")
    print("Accuracy:", accuracy_score(y_test, pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, pred))
    print("Classification Report:\n", classification_report(y_test, pred))
    print()
```

```python
from sklearn.model_selection import GridSearchCV

knn_params = {
    'n_neighbors': list(range(3, 15)),
    'weights': ['uniform', 'distance']
}

knn_grid = GridSearchCV(KNeighborsClassifier(), knn_params, cv=5, scoring='accuracy')
knn_grid.fit(X_train, y_train)

print("Best KNN Params:", knn_grid.best_params_)
knn_best = knn_grid.best_estimator_
```

```python
svm_params = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 0.01, 0.001],
    'kernel': ['rbf', 'linear']
}

svm_grid = GridSearchCV(SVC(probability=True), svm_params, cv=5, scoring='accuracy')
svm_grid.fit(X_train, y_train)

print("Best SVM Params:", svm_grid.best_params_)
svm_best = svm_grid.best_estimator_
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Get probabilities
rf_probs = rf.predict_proba(X_test)[:, 1]
knn_probs = knn_best.predict_proba(X_test)[:, 1]
svm_probs = svm_best.predict_proba(X_test)[:, 1]

# Compute ROC
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_probs)
fpr_knn, tpr_knn, _ = roc_curve(y_test, knn_probs)
fpr_svm, tpr_svm, _ = roc_curve(y_test, svm_probs)

# AUC
auc_rf = auc(fpr_rf, tpr_rf)
auc_knn = auc(fpr_knn, tpr_knn)
auc_svm = auc(fpr_svm, tpr_svm)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc_rf:.2f})')
plt.plot(fpr_knn, tpr_knn, label=f'KNN (AUC = {auc_knn:.2f})')
plt.plot(fpr_svm, tpr_svm, label=f'SVM (AUC = {auc_svm:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid()
plt.show()
```
